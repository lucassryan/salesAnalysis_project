```sql
--==========================================
-- Verificando a estrutura de cada tabela --

SELECT * FROM sales_project.olist_ecommerce.olist_orders_dataset LIMIT 10;
SELECT * FROM sales_project.olist_ecommerce.olist_order_items_dataset LIMIT 10;
SELECT * FROM sales_project.olist_ecommerce.olist_customers_dataset LIMIT 10;
SELECT * FROM sales_project.olist_ecommerce.olist_products_dataset LIMIT 10;
```

```sql
--===============================================
-- Contando o número de linhas de cada tabela --
SELECT COUNT(*) FROM sales_project.olist_ecommerce.olist_orders_dataset;
SELECT COUNT(*) FROM sales_project.olist_ecommerce.olist_order_items_dataset;
SELECT COUNT(*) FROM sales_project.olist_ecommerce.olist_customers_dataset;
SELECT COUNT(*) FROM sales_project.olist_ecommerce.olist_products_dataset;
```

```sql
--=================================
--Verificando a chave de cada tabela

-- sales_project.olist_ecommerce.olist_orders_dataset
SELECT
  COUNT(*) AS total_linhas,
  COUNT(DISTINCT order_id) AS pedidos_unicos
FROM sales_project.olist_ecommerce.olist_orders_dataset;


-- sales_project.olist_ecommerce.olist_customers_dataset
SELECT
    COUNT(*) AS total_linhas,
    COUNT(DISTINCT customer_id) AS clientes_unicos
FROM sales_project.olist_ecommerce.olist_customers_dataset;


--sales_project.olist_ecommerce.olist_products_dataset
SELECT
    COUNT(*) AS total_linhas,
    COUNT(DISTINCT product_id) AS produtos_unicos
FROM sales_project.olist_ecommerce.olist_products_dataset;


--sales_project.olist_ecommerce.olist_order_items_dataset
SELECT
    COUNT(*) AS total_linhas,
    COUNT(DISTINCT order_id, order_item_id) AS itens_unicos
FROM sales_project.olist_ecommerce.olist_order_items_dataset;

```
